```python
# Normal Equation soulution
import os, pandas as pd
import numpy as np
csv_path = os.path.join("./", "allPeople.csv")
people = pd.read_csv(csv_path)
people.plot(kind='scatter', x="weight", y='height', figsize=(5,3))

X = people['weight']
y = people['height']

X_b = np.c_[np.ones((len(X), 1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
X_new = np.array([[0], [120]])
X_new_b = np.c_[np.ones((2, 1)), X_new] 
y_predict = X_new_b.dot(theta_best)
plt.axis([30, 120, 100, 310])

plt.plot(X_new, y_predict, "r-")
print('Theta = ' + str(theta_best))
plt.show()
```

```python
# Gradient Descent
import os, pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_path = os.path.join("./", "allPeople.csv")
people = pd.read_csv(csv_path)
people.plot(kind='scatter', x="weight", y='height', figsize=(5,3))

X = people['weight']
y = people['height']
X = np.c_[X]
y = np.c_[y]

X_b = np.c_[np.ones((len(X), 1)), X]

eta = 0.0001
n_iterations = 100000
m = len(X)
theta = np.random.randn(2,1)

for i in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta)-y)
    theta = theta - eta * gradients

X_new = np.array([[0], [120]])
X_new_b = np.c_[np.ones((2, 1)), X_new] 
y_predict = X_new_b.dot(theta)
plt.axis([30, 120, 100, 310])

plt.plot(X_new, y_predict, "r-")
print('Theta = ' + str(theta))
plt.show()
```
